```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# Create a data frame with the personal information of each player

players_infodf = purchase_data[['SN', 'Age', 'Gender']]

# Drop duplicates
players_infodf.drop_duplicates(subset="SN", keep = 'first', inplace = True)

# Update the SN header to User Name
players_infodf = players_infodf.rename(columns = {"SN": "User Name"})
```

```python
# Count the number of players
players_count = players_infodf["User Name"].count()
```

```python
# Creating a data frame using the len of SN to visualize the total of players
total_players = pd.DataFrame({"Total Players": players_count}, index=[0])
total_players.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Obtain unique items
unique_items = len(purchase_data["Item Name"].unique())
```

```python
# Average Price
average_price = round((purchase_data["Price"].mean()),2)
```

```python
# Number of purchases
total_purchases = purchase_data["Purchase ID"].count()
```

```python
# Total Revenue
total_revenue = purchase_data["Price"].sum()
```

```python
# Creating a summary data frame to hold the results + currency formatting
purchase_analysis = pd.DataFrame({"Number of Unique Items": [unique_items],
                                   "Average Price": [average_price],
                                   "Number of Purchases": [total_purchases],
                                   "Total Revenue": [total_revenue]}).style.format({"Average Price":"${:,.2f}",
                                   "Total Revenue":"${:,.2f}"})
```

```python
# Display the summary data frame
purchase_analysis
```
